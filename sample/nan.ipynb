```python
# setting
import pandas as pd
from pandasql import sqldf
data = pd.read_csv('./data.csv')
```

#### 특정 컬럼에 존재하는 결측값 확인

```python
# pandas
data['rating'].isnull().sum()
```

```python
# SQL
query = \
'''
SELECT COUNT(CASE WHEN rating IS NULL THEN 1 
             END) as rating_null
FROM data 
'''
sqldf(query)
```

#### 특정 컬럼에서 결측값이 존재하는 경우 확인

```python
# pandas
data[data['rating'].isna()]
```

```python
# SQL
query = \
'''
SELECT * 
FROM data
WHERE rating IS NULL 
'''
sqldf(query)
```

#### 특정 컬럼에서 결측값이 없는 경우 확인

```python
# pandas
data[~data['rating'].isna()]
```

```python
# SQL
query = \
'''
SELECT * 
FROM data
WHERE rating IS NOT NULL 
'''
sqldf(query)
```
